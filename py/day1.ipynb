````python
# `ndim`表示数组维度(轴)的个数，也称为数组的秩
# `shape`表示数组的形状，它是一个整数元组，长度等于`ndim`
arr_1_d = np.array([1])
print('arr_1_d:', arr_1_d)
print('arr_1_d.ndim:', arr_1_d.ndim)
print('arr_1_d.shape:', arr_1_d.shape)
print('arr_1_d.size:', arr_1_d.size)
print('arr_1_d.dtype:', arr_1_d.dtype)
print('---------------------------------')
arr_1_d = np.arange(6)
print('arr_1_d:', arr_1_d)
print('arr_1_d.ndim:', arr_1_d.ndim)
print('arr_1_d.shape:', arr_1_d.shape)
print('arr_1_d.size:', arr_1_d.size)
print('arr_1_d.dtype:', arr_1_d.dtype)
print('---------------------------------')
arr_2_d = np.arange(6).reshape(2,3)
print('arr_2_d:', arr_2_d)
print('arr_2_d.ndim:', arr_2_d.ndim)
print('arr_2_d.shape:', arr_2_d.shape)
print('arr_2_d.size:', arr_2_d.size)
print('arr_2_d.dtype:', arr_2_d.dtype)
print('---------------------------------')
arr_3_d = np.arange(6).reshape(1, 3, 2)
print('arr_3_d:', arr_3_d)
print('arr_3_d.ndim:', arr_3_d.ndim)
print('arr_3_d.shape:', arr_3_d.shape)
print('arr_3_d.size:', arr_3_d.size)
print('arr_3_d.dtype:', arr_3_d.dtype)
arr_3_d.reshape(2,3)
````

````python
# 4名同学对3款游戏的评分
scores = np.random.randint(10, size=(4, 3))
print('scores:', scores)
print('scores.ndim:', scores.ndim)
print('scores.shape:', scores.shape)
print('scores.size:', scores.size)
# axis=i表示沿着第i个轴进行操作, 第i个轴的维度会被消除(折叠或聚合到一起)
# 每款游戏的总评分
sum0 = np.sum(scores, axis=0)
print('sum0: ', sum0, sum0.shape)
# 每名同学的总评分
sum1 = np.sum(scores, axis=1)
print('sum1: ', sum1, sum1.shape)
# 每名同学打分的平均值
avg1 = np.average(scores, axis=1)
print('avg1: ', avg1)
# 每款游戏的平均分
avg0 = np.average(scores, axis=0)
print('avg0: ', avg0)
````

````python
from PIL import Image
import numpy as np
import cv2

im_pil = Image.open('jk.webp')
print(type(im_pil))
print(im_pil.format, im_pil.size, im_pil.mode)

arr_pil = np.asarray(im_pil)
print('arr_pil:', arr_pil.shape, arr_pil.ndim, arr_pil.dtype)

im_cv = cv2.imread('jk.webp')
print(type(im_cv))
arr_cv = np.asarray(im_cv)
print('arr_cv:', arr_cv.shape, arr_cv.ndim, arr_cv.dtype)

arr_pil_c1 = arr_pil[:, :, 0]
arr_pil_c1 = arr_pil_c1[:, :, np.newaxis]
print('arr_pil_c1:', arr_pil_c1.shape, arr_pil_c1.ndim, arr_pil_c1.dtype)
arr_pil_c2 = arr_pil[:, :, 1]
arr_pil_c2 = arr_pil_c2[:, :, np.newaxis]
arr_pil_c3 = arr_pil[:, :, 2]
print('arr_pil_c3:', arr_pil_c3.shape, arr_pil_c3.ndim)

zeros = np.zeros(arr_pil_c1.shape, dtype=np.uint8)
print('zeros:', zeros.shape, zeros.ndim)

arr_pil_c1_3ch = np.concatenate((arr_pil_c1, zeros, zeros), axis=2)
print('arr_pil_c1_3ch:', arr_pil_c1_3ch.shape, arr_pil_c1_3ch.ndim)
arr_pil_c2_3ch = np.concatenate((zeros, arr_pil_c2, zeros), axis=2)
arr_pil_c3_3ch = np.zeros(arr_pil.shape, dtype=np.uint8)
arr_pil_c3_3ch[:, :, 2] = arr_pil_c3
print('arr_pil_c3_3ch:', arr_pil_c3_3ch.shape, arr_pil_c3_3ch.ndim)

from matplotlib import pyplot as plt
plt.figure()
plt.subplot(2, 2, 1)
plt.title('Original')
plt.imshow(arr_pil)
plt.axis('off')
plt.subplot(2, 2, 2)
plt.title('PIL C1')
plt.imshow(arr_pil_c1_3ch)
plt.axis('off')
plt.subplot(2, 2, 3)
plt.title('PIL C2')
plt.imshow(arr_pil_c2_3ch)
plt.axis('off')
plt.subplot(2, 2, 4)
plt.title('PIL C3')
plt.imshow(arr_pil_c3_3ch)
plt.axis('off')

arr_cv_c1_3ch = np.zeros(arr_cv.shape, dtype=np.uint8)
arr_cv_c1_3ch[:, :, 0] = arr_cv[:, :, 2]
arr_cv_c2_3ch = np.zeros(arr_cv.shape, dtype=np.uint8)
arr_cv_c2_3ch[:, :, 1] = arr_cv[:, :, 1]
arr_cv_c3_3ch = np.zeros(arr_cv.shape, dtype=np.uint8)
arr_cv_c3_3ch[:, :, 2] = arr_cv[:, :, 0]
print('arr_cv_c3_3ch:', arr_cv_c3_3ch.shape, arr_cv_c3_3ch.ndim)
rgb = np.zeros(arr_cv.shape, dtype=np.uint8)
rgb[:, :, 0] = arr_cv[:, :, 2]
rgb[:, :, 1] = arr_cv[:, :, 1]
rgb[:, :, 2] = arr_cv[:, :, 0]
print('rgb:', rgb.shape, rgb.ndim)
plt.figure()
plt.subplot(2, 2, 1)
plt.title('Original')
plt.imshow(rgb)
plt.axis('off')
plt.subplot(2, 2, 2)
plt.title('CV C1')
plt.imshow(arr_cv_c1_3ch)
plt.axis('off')
plt.subplot(2, 2, 3)
plt.title('CV C2')
plt.imshow(arr_cv_c2_3ch)
plt.axis('off')
plt.subplot(2, 2, 4)
plt.title('CV C3')
plt.imshow(arr_cv_c3_3ch)
plt.axis('off')
````
